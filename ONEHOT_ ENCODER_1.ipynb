```python
import cv2
import os
import random
import numpy as np

# Tentukan jalur direktori gambar asli dan direktori tempat gambar hasil augmentasi akan disimpan
original_path = r"C:\Users\lenovo\Downloads\TA\Refuge dataset\DATA ORI IMAGE\Severe Glaucoma"
augmented_path = r"C:\Users\lenovo\Downloads\TA\Refuge dataset\DATA TRAIN PREPRO CROP\SEVERE"

# Fungsi untuk meningkatkan kualitas gambar dan meresizem
def enhance_and_resize_image(image, output_size=(256, 256)):
    # Sharpening menggunakan Unsharp Masking
    gaussian_blur = cv2.GaussianBlur(image, (9, 9), 10.0)
    unsharp_image = cv2.addWeighted(image, 1.5, gaussian_blur, -0.5, 0)

    # Histogram Equalization (CLAHE)
    lab = cv2.cvtColor(unsharp_image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    cl = clahe.apply(l)
    limg = cv2.merge((cl, a, b))
    enhanced_image = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

    # Resize image
    resized_image = cv2.resize(enhanced_image, output_size)

    return resized_image

# Fungsi untuk melakukan augmentasi pada gambar
def augment_image(image, output_size=(256, 256)):
    augmented_images = []

    # Terapkan filter untuk meningkatkan kualitas gambar dan resize
    enhanced_image = enhance_and_resize_image(image, output_size)

    # Penyesuaian kontras dan kecerahan
    for _ in range(2): 
        alpha = random.uniform(0.8, 1.2)  # Faktor kontras secara acak antara 0.8 dan 1.2
        beta = random.randint(-30, 30)  # Nilai kecerahan secara acak antara -30 dan 30
        adjusted_image = cv2.convertScaleAbs(enhanced_image, alpha=alpha, beta=beta)
        augmented_images.append(adjusted_image)

    # Rotasi tetap 90, 180, dan 270 derajat
    rotated_image_90 = cv2.rotate(enhanced_image, cv2.ROTATE_90_CLOCKWISE)
    rotated_image_180 = cv2.rotate(enhanced_image, cv2.ROTATE_180)
    rotated_image_270 = cv2.rotate(enhanced_image, cv2.ROTATE_90_COUNTERCLOCKWISE)
    
    augmented_images.extend([rotated_image_90, rotated_image_180, rotated_image_270])

    return augmented_images

# Buat direktori untuk gambar hasil augmentasi jika belum ada
if not os.path.exists(augmented_path):
    os.makedirs(augmented_path)

# Counter untuk gambar hasil augmentasi
total_augmented_images = 0
max_augmented_images = 400 # Jumlah gambar augmented yang diinginkan

# Iterate melalui gambar-gambar asli dan lakukan augmentasi
original_images = [f for f in os.listdir(original_path) if f.endswith('.jpg') or f.endswith('.png')]

while total_augmented_images < max_augmented_images:
    for filename in original_images:
        if total_augmented_images >= max_augmented_images:
            break
        image_path = os.path.join(original_path, filename)
        image = cv2.imread(image_path)

        # Augmentasi gambar
        augmented_images = augment_image(image, output_size=(256, 256))

        # Simpan gambar hasil augmentasi
        base_filename = os.path.splitext(filename)[0]
        for idx, augmented_image in enumerate(augmented_images):  
            if total_augmented_images >= max_augmented_images:
                break
            augmented_filename = f'{base_filename}_aug_{total_augmented_images}.jpg'
            augmented_image_path = os.path.join(augmented_path, augmented_filename)
            cv2.imwrite(augmented_image_path, augmented_image)
            total_augmented_images += 1

print('Augmentasi selesai.')
```

```python
import cv2
import os
import random
import numpy as np

# Tentukan jalur direktori gambar asli dan direktori tempat gambar hasil augmentasi akan disimpan
original_path = r"C:\Users\lenovo\Downloads\TA\ORIGA DATASET\image cropped class\IMAGE CROP ORI\normal"
augmented_path = r"C:\Users\lenovo\Downloads\TA\ORIGA DATASET\image cropped class\IMAGE CROP NON AUGMENTED BUT SHARPENED\NORMAL"

# Fungsi untuk meningkatkan kualitas gambar
def enhance_image(image):
    # Sharpening menggunakan Laplacian Filter
    laplacian_filter = np.array([[0, -1, 0], [-1, 5,-1], [0, -1, 0]])
    sharpened_image = cv2.filter2D(image, -1, laplacian_filter)

    # Histogram Equalization (CLAHE)
    lab = cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    cl = clahe.apply(l)
    limg = cv2.merge((cl, a, b))
    enhanced_image = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

    return enhanced_image

# Fungsi untuk melakukan augmentasi pada gambar
def augment_image(image):
    augmented_images = []

    # Terapkan filter untuk meningkatkan kualitas gambar
    enhanced_image = enhance_image(image)

    # Penyesuaian kontras dan kecerahan
    for _ in range(2): 
        alpha = random.uniform(0.8, 1.2)  # Faktor kontras secara acak antara 0.8 dan 1.2
        beta = random.randint(-30, 30)  # Nilai kecerahan secara acak antara -30 dan 30
        adjusted_image = cv2.convertScaleAbs(enhanced_image, alpha=alpha, beta=beta)
        augmented_images.append(adjusted_image)

    # Rotasi tetap 90, 180, dan 270 derajat
    rotated_image_90 = cv2.rotate(enhanced_image, cv2.ROTATE_90_CLOCKWISE)
    rotated_image_180 = cv2.rotate(enhanced_image, cv2.ROTATE_180)
    rotated_image_270 = cv2.rotate(enhanced_image, cv2.ROTATE_90_COUNTERCLOCKWISE)
    
    augmented_images.extend([rotated_image_90, rotated_image_180, rotated_image_270])

    return augmented_images

# Buat direktori untuk gambar hasil augmentasi jika belum ada
if not os.path.exists(augmented_path):
    os.makedirs(augmented_path)

# Counter untuk gambar hasil augmentasi
total_augmented_images = 0
max_augmented_images = 100  # Jumlah gambar augmented yang diinginkan

# Iterate melalui gambar-gambar asli dan lakukan augmentasi
original_images = [f for f in os.listdir(original_path) if f.endswith('.jpg') or f.endswith('.png')]

while total_augmented_images < max_augmented_images:
    for filename in original_images:
        if total_augmented_images >= max_augmented_images:
            break
        image_path = os.path.join(original_path, filename)
        image = cv2.imread(image_path)

        # Augmentasi gambar
        augmented_images = augment_image(image)

        # Simpan gambar hasil augmentasi
        base_filename = os.path.splitext(filename)[0]
        for idx, augmented_image in enumerate(augmented_images):  
            if total_augmented_images >= max_augmented_images:
                break
            augmented_filename = f'{base_filename}_aug_{total_augmented_images}.jpg'
            augmented_image_path = os.path.join(augmented_path, augmented_filename)
            cv2.imwrite(augmented_image_path, augmented_image)
            total_augmented_images += 1

print('Augmentasi selesai.')
```

```python
import cv2
import numpy as np

# Fungsi untuk menghitung diameter dari kontur
def calculate_diameter(contour):
    x, y, w, h = cv2.boundingRect(contour)
    diameter = max(w, h)
    return diameter

# Load image
image = cv2.imread(r"C:\Users\DELL\Downloads\segmentation_results_origa\570_overlay.png", cv2.IMREAD_GRAYSCALE)

# Thresholding untuk segmentasi
_, thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Temukan kontur
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Misalkan kontur terbesar adalah disk dan kontur kedua terbesar adalah cup
contours = sorted(contours, key=cv2.contourArea, reverse=True)
disk_contour = contours[0]
cup_contour = contours[1]

# Hitung diameter dari cup dan disk
disk_diameter = calculate_diameter(disk_contour)
cup_diameter = calculate_diameter(cup_contour)

# Hitung CDR
cdr = cup_diameter / disk_diameter

print("CDR:", cdr)
```
